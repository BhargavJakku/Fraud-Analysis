# Data Wrangling

# checking if any columns contains maximum value equal to 0 and dropping that

```python
zero_features = (df.max() == 0)
zero_features = zero_features[zero_features == True].index
print("Dropping the features which have maximum value 0:\n", zero_features)
df=df.drop(zero_features, axis=1)
```

# Since asn_ip is a unique id, we can drop this attribute. 

```python
df=df.drop(['asn_ip'],axis=1)
```

# missing values is represented by -1 so we can replace it with nan

```python
import numpy as np
df = df.replace(-1, np.nan)
df.head()
```

```python
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})
missing_value_df
```

# observation:
Almost all the features contain missing values more than 50% which means most of the values in the data frame are empty.for that we need to drop some of the feature and for remaining features we need to drop those data points based on rows.

# dropping the features that contain more that 50% missing values

```python

missing = df.isnull().sum()/len(df)
features = missing[missing > 0.5].index
print("Dropping the variables that have more than 50% missing values:\n")
print(features)
```

```python
df=df.drop(features, axis=1)
df
```

# dropping missing values based on rows

```python
df=df.dropna(axis=0)
```

```python
#make plot function and set parameters
def make_plots(feature, title="", limited=False, n=10):
    print("Total unique values are: ", len(feature.value_counts()), "\n\n")
    print("Category\tValue\n")
    if limited:
        data = feature.value_counts()[0:n]
    else:
        data = feature.value_counts()
    print(data)
    categories_num = len(data)
    #plotting bar-plot and pie chart
    sns.set_style('darkgrid')
    #Set fig size
    plt.figure(figsize=(16,5))
    #ploting subplot
    plt.subplot(1,2,1)
    plt.title(title, fontsize=16)
    plt.xticks(rotation=45)
    #ploting barplot
    plot = sns.barplot(x=data.index, y=data.values, edgecolor="white", palette=sns.palettes.color_palette("icefire"))
    total = len(feature)
    for p in plot.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.08
        y = p.get_y() + p.get_height()
        plot.annotate(percentage, (x, y), size = 12)
    
    plt.subplot(1,2,2)
    labels = data.index
    plt.pie(x=data, autopct="%.1f%%", explode=[0.02]*categories_num, labels=labels, pctdistance=0.5)
    plt.title(title, fontsize=16)
    plt.show()
```

```python
#calling make_plot function
make_plots(df["phishing"],"Distribution of phishing", limited=True, n=15)
```

# data preprocessing

# defining input and output as x and y respectively

```python
X=df.drop(['phishing'],axis=1)
y=df['phishing']
```

# Ovsersampling using smote
As we have seen from the above analysis that our dataset is highly imbalanced so we decided to balance the data using smote oversampling technique

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from collections import Counter
over = SMOTE(random_state=1000,sampling_strategy=0.8)
under = RandomUnderSampler(random_state=1000,sampling_strategy=0.6)
steps = [('o', under), ('u', over)]
pipeline = Pipeline(steps=steps)
# transform the dataset
X, y = pipeline.fit_resample(X, y)
# summarize class distribution
print(Counter(y))
```

# Test train split of data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

# DecisionTreeClassifier

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
param_grid = {'max_features': ['auto', 'sqrt', 'log2'],
              'ccp_alpha': [0.1, .01, .001],
              'max_depth' : [5, 6, 7, 8, 9],
              'criterion' :['gini', 'entropy']
             }
tree_clas = DecisionTreeClassifier(random_state=1024)
grid_search = GridSearchCV(estimator=tree_clas, param_grid=param_grid, cv=5, verbose=True)
grid_search.fit(X_train, y_train)
```

```python
final_model = grid_search.best_estimator_
final_model
```

```python
from sklearn.metrics import confusion_matrix

dt =DecisionTreeClassifier(ccp_alpha=0.001, criterion='entropy', max_depth=8,
                       max_features='auto', random_state=1024)
dt.fit(X_train, y_train)
y_predict_dt = dt.predict(X_test)
# confusion_matrix
plt.figure(figsize=(10,10))
cm = confusion_matrix(y_test, y_predict_dt)
sns.heatmap(cm, annot=True, fmt=".2f")
```

```python
print('accuracy',accuracy_score((y_test), y_predict_dt))
print('precision',precision_score(y_test, y_predict_dt, average='macro'))
print('recall',recall_score(y_test, y_predict_dt, average='macro'))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
import warnings
warnings.filterwarnings('ignore')

params = {'n_neighbors':range(1,20),"metric":["euclidean", "manhattan"],
              'algorithm' :['ball_tree','kd_tree','brute']
        
             }
knn=KNeighborsClassifier()
gs = GridSearchCV(knn, param_grid= params, cv = 5,scoring = "roc_auc",verbose=1)
gs.fit(X_train, y_train)
print(gs.best_score_)
print(gs.best_estimator_)
print(gs.best_params_)
```

# ROC Curve for all models

```python
### Compute ROC curve and ROC area for predictions on validation set
from sklearn.metrics import roc_curve, auc
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_predict_lr)
roc_auc_lr = auc(fpr_lr, tpr_lr)
### Plot
plt.figure(figsize=(15,15))
lw = 2
plt.plot(fpr_lr, tpr_lr, color='red',
         lw=lw, label='Logistic regression (area = %0.2f)' % roc_auc_lr)
fpr_knn, tpr_knn, _ = roc_curve(y_test, y_predict_knn)
roc_auc_knn = auc(fpr_knn, tpr_knn)
plt.plot(fpr_knn, tpr_knn, color='darkorange',
         lw=lw, label='knn (area = %0.2f)' % roc_auc_knn)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_predict_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)
plt.plot(fpr_rf, tpr_rf, color='blue',
         lw=lw, label='Random Forest (area = %0.2f)' % roc_auc_rf)
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_predict_dt)
roc_auc_dt = auc(fpr_dt, tpr_dt)
plt.plot(fpr_dt, tpr_dt, color='green',
         lw=lw, label='Decision Tree (area = %0.2f)' % roc_auc_dt)

plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
